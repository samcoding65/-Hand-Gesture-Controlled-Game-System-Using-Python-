```python
import cv2
import mediapipe as mp
import pyautogui
# import math
import time

cam=cv2.VideoCapture(0)
mp_hands=mp.solutions.hands
hands=mp_hands.Hands(static_image_mode=False,
    max_num_hands=1,
    min_detection_confidence=0.7,
    min_tracking_confidence=0.7)
draw=mp.solutions.drawing_utils

last_pressed_time = 0
cooldown = 1.0

while True:
    sucess,frame=cam.read()
    frame=cv2.flip(frame,1)
    h, w, _ = frame.shape
    frame_rgb=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    results=hands.process(frame_rgb)

    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            draw.draw_landmarks(frame,hand_landmarks,mp_hands.HAND_CONNECTIONS)

            landmarks=hand_landmarks.landmark
            

            finger_up=0
            finger_ids=[(8,6),(12,10),(16,14),(20,18)]

            index_fing=landmarks[8].y < landmarks[6].y
            middle_fing=landmarks[12].y < landmarks[10].y
            ring_fing=landmarks[16].y < landmarks[14].y
            pinky_fing=landmarks[20].y < landmarks[18].y
            ring_down = landmarks[16].y > landmarks[14].y
            pinky_down = landmarks[20].y > landmarks[18].y

            for tip,pip in finger_ids:
                if landmarks[tip].y<landmarks[pip].y:
                    finger_up+=1

            if finger_up==4:
                current_time = time.time()
                if current_time - last_pressed_time > cooldown:
                    pyautogui.press('up')
                    print('up pressed')
                    last_pressed_time = current_time
                    


            if ( index_fing and not(middle_fing or ring_fing or pinky_fing)):
                current_time = time.time()
                if current_time - last_pressed_time > cooldown:
                    pyautogui.press('left')
                    print('LEFT done')
                    last_pressed_time = current_time

            if( pinky_fing):
                current_time = time.time()
                if current_time - last_pressed_time > cooldown:
                    pyautogui.press('right')
                    print('RIGHT')
                    last_pressed_time = current_time

            if(index_fing and middle_fing and ring_down and pinky_down):
                current_time = time.time()
                if current_time - last_pressed_time > cooldown:
                    pyautogui.press('down')
                    print('DOWN')
                    last_pressed_time = current_time


    cv2.imshow("hand gesture",frame)
    if cv2.waitKey(1)&0xFF==ord('q'):
        break

cam.release()
cv2.destroyAllWindows()
```
